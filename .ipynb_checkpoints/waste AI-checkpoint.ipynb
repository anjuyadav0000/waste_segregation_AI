```python
import zipfile

with zipfile.ZipFile("dataset/Trashnet.zip", "r") as zip_ref:
    zip_ref.extractall("dataset")

print("Dataset Extracted Successfully!")
```

```python
import os

print(os.listdir("dataset/dataset-resized"))
```

```python
import os

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Flatten, Dense

# Create model folder
os.makedirs("model", exist_ok=True)

print("Loading Dataset...")

# Data Generator
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

# Training Data
train = datagen.flow_from_directory(
    "dataset/dataset-resized",
    target_size=(224,224),
    batch_size=32,
    class_mode="categorical",
    subset="training"
)

# Validation Data
val = datagen.flow_from_directory(
    "dataset/dataset-resized",
    target_size=(224,224),
    batch_size=32,
    class_mode="categorical",
    subset="validation"
)

print("Building Model...")

# CNN Model
model = Sequential([

    Conv2D(
        32,
        (3,3),
        activation="relu",
        input_shape=(224,224,3)
    ),
    MaxPooling2D(),

    Conv2D(
        64,
        (3,3),
        activation="relu"
    ),
    MaxPooling2D(),

    Flatten(),

    Dense(
        128,
        activation="relu"
    ),

    Dense(
        6,
        activation="softmax"
    )

])

# Compile Model
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

print("Training Started...")

# Train Model
history = model.fit(
    train,
    validation_data=val,
    epochs=10
)

print("Saving Model...")

# Save Model
model.save("model/waste_model.keras")

print(" Model Saved Successfully!")
print(" Saved at: model/waste_model.keras")
```
